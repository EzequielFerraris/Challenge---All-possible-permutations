FIRST SOLUTION

1. We have to come to a form of combining elements in all the possible ways. But the problem is that "all the possible ways" does include multiple combinations where an element is at the same position while others elements change their position. So the solution does not consist in iterating each element n number of times. What is important is their relative position.
2. So we can preserve the link between elements reducing the unity to pairs. If we have all the possible pairs for the list, with repeated elements, we can have all the possible positions that the elements can have between them. 
3. First we build a function that iterates over the original list and returns a list with all the possible pairs.
4. Then we build a function that uses that list of pairs to combine recursively those pairs in full array of elements. 
5. To do this, we first return the passed list if it has length = 1. An we create a variable where we can store the permutations. This will be the final output of the function.
6. Then we create a function inside that we will call recursively. It will receive only one list with a pair of values as a parameter. We limit the function to return a result if the array in construction reaches the length of the original list. This is the ending clause.
6. We create a list that includes all the pairs that start with the last element of the passed array and which its second element is not already in the array.
7. We create a second list with length equal to the length of the previous list, but which has elements copies of the passed array.
8. Then we append the last element of each of the pairs in the first created list to each copy of the original array.
9. Finally, we call this function recursively for each element of the second list. It will generate all the possible permutations for the first pair passed as paramether.
10. Outside the recursive function, we iterate over all the possible pairs aplying the recursive function. 
11. We return the array containing all the possible permutations. 

```python
#MAKE PAIRS FROM THE ELEMENTS OF THE LIST
def generate_pairs(list1):
    pairs = []
    for element in list1:
        for element2 in range(len(list1)):
            if element == list1[element2]:
                continue
            else:
                pairs.append([element, list1[element2]])
    return pairs


def permutations(list1):

    if len(list1) == 1:
        return list1

    pairs = generate_pairs(list1)
    
    permutations=[]

    def recursive(p2):
        
        if len(p2) == len(list1):
            return permutations.append(p2)

        a_list = [_[1] for _ in pairs if (_[0] == p2[-1]) and (_[1] not in p2)]

        nl = [[_ for _ in p2] for x in range(len(a_list))]
        
        for _ in range(len(a_list)):
            nl[_].append(a_list[_])
        
        for element in nl:
            recursive(element)

    for pair in pairs:
        recursive(pair)
        
    return permutations




```
